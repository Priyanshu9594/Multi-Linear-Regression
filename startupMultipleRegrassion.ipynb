```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import statsmodels.formula.api as smf
```

```python
startup=pd.read_csv("C:/Users/user/Downloads/50_Startups.csv")
startup.head()
```

```python
startup1=startup.rename({'R&D Spend':'RandD'}, axis=1)

startup1
```

```python
startup2=startup1.rename({'Marketing Spend':'MarketingSpend'}, axis=1)
startup2.head()
```

```python
startup2.info()
```

```python
startup2.corr()
```

```python
sns.set_style(style='whitegrid')
sns.pairplot(startup2)
```

```python
import statsmodels.formula.api as smf 
model=smf.ols('RandD ~ Administration+MarketingSpend+Profit', data=startup2).fit()
model
```

```python
model.params
```

```python
#t and p-Values
print(model.tvalues, '\n', model.pvalues)
```

```python
#R squared values
(model.rsquared,model.rsquared_adj)
```
